```python
df2['Distance'] = df2['Event distance/length'].str.extract(r'([\d.]+)').astype(float)  # Extracts the numerical part
df2['Unit'] = df2['Event distance/length'].str.extract(r'([a-zA-Z]+)')  # Extracts the unit part

# Convert mi to km
conversion_factor = 1.60934
df2.loc[df2['Unit'] == 'mi', 'Distance'] *= conversion_factor

df2['Event length'] = df2['Distance'].round().astype(int).astype(str) + 'km'

df2.drop(columns=['Distance', 'Unit'], inplace=True)
```

```python
age_grp=df2[df2['Event distance/length']=="50mi"].groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False).query('count>20')
age_grp
```

```python
age_grp.head(10).plot(kind='bar',figsize=(12,6),title='Age Wise Mean Average Speed and Participants Count ')
```

```python
def age_grp(x):
    if(x<20):
        return 'Less than 20'
    elif (x>=20 and x<=30):
        return '20 to 30'
    elif (x>30 and x<=40):
        return '31 to 40'
    elif (x>40 and x<=50 ):
        return '41 to 50'
    else :
        return 'upper 50'
    


df2['age_cat']=df2['athlete_age'].apply(age_grp)
```

```python
ag=df2.age_cat.value_counts()
ag
```
