```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combine all filters together
```

```python
df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
#remove (USA) from event name
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#clean up  athelete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#Remove the h from the 'Athlete Performance' to show the hours
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
#drop columns: Athlete club, Athlete Country,Athlete year of birth
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth'], axis = 1)
```

```python
df2.head()
```

```python
#clean up null values
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 =df2.dropna()
```

```python
df2.shape
```

```python
#check  for  dupes
```

```python
df2[df2.duplicates() == True]
```

```python
#reset index
```

```python
df2.reset_index(drop=True)
```

```python
df2.dtypes
```

```python
df2['Distance'] = df2['Event distance/length'].str.extract(r'([\d.]+)').astype(float)  # Extracts the numerical part
df2['Unit'] = df2['Event distance/length'].str.extract(r'([a-zA-Z]+)')  # Extracts the unit part

# Convert mi to km
conversion_factor = 1.60934
df2.loc[df2['Unit'] == 'mi', 'Distance'] *= conversion_factor

df2['Event length'] = df2['Distance'].round().astype(int).astype(str) + 'km'

df2.drop(columns=['Distance', 'Unit'], inplace=True)
```

```python
df2.head()
```

```python
df2.reset_index(drop=True)
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Average athlete_age'] = df2['athlete_age'].astype(float)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
#rename column names
```

```python
df2 = df2.rename(columns = {'Year of event' : 'year',
                                    'Event dates': 'race_day',                  
                                    'Event name'       : 'race_name',                    
                                    'Event number of finishers'  : 'race_number_of_finishers',     
                                    'Athlete performance'  : 'athlete_performance',          
                                    'Athlete gender' : 'athlete_gender',                
                                    'Athlete age category'  : 'athlete_age_category',         
                                    'Athlete average speed'  : 'athlete_average_speed',        
                                    'Athlete ID'    : 'athlete_id',
                                    'Event length' : 'race_length'})
```
